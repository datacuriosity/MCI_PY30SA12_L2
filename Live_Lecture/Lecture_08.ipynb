```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 1000)

y = sigmoid(x)

plt.plot(x, y, color='r', label='Sigmoid')
plt.hlines(y=0.5, xmin=-10, xmax=10, linestyles='--')
plt.vlines(x=0, ymin=0, ymax=1, linestyles='-.')
plt.legend(loc='upper left');
```

**Cancer or not**
- Cancer: 1  --> sigmoid output
- Not Cancer: 0
